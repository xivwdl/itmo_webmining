### Задача 2
Вася составил раскладку по продуктам на один день (она на листе "Раскладка") с указанием названия продукта и его количества в граммах. Посчитайте 4 числа: суммарную калорийность и граммы белков, жиров и углеводов. Числа округлите до целых вниз и введите через пробел

```python
# Подключаем файл и заменяем NaN на 0, как в условии задачи
# Изменяем имя 1 столбца в справочнике для удобства
df_dir = pd.read_excel('trekking2.xlsx', sheet_name='Справочник').fillna(0)
df_dir.rename(columns={'Unnamed: 0':'Продукт'}, inplace=True)
df_day = pd.read_excel('trekking2.xlsx', sheet_name='Раскладка')

print(df_dir.head(10))
print(df_day)
```

```python
# Добавляем недостающую информацию из справочника в раскладку
df_day = df_day.merge(df_dir, on='Продукт')
df_day
```

```python
# Проводим необходимые подсчеты
df_day['Кол-во ККал'] = df_day['ККал на 100'] / 100 * df_day['Вес в граммах']
df_day['Кол-во Б'] = df_day['Б на 100'] / 100 * df_day['Вес в граммах']
df_day['Кол-во Ж'] = df_day['Ж на 100'] / 100 * df_day['Вес в граммах']
df_day['Кол-во У'] = df_day['У на 100'] / 100 * df_day['Вес в граммах']

print('Суммарная калорийность:', int(df_day['Кол-во ККал'].sum()))
print('Сумма белков в граммах:', int(df_day['Кол-во Б'].sum()))
print('Сумма жиров в граммах:', int(df_day['Кол-во Ж'].sum()))
print('Сумма углеводов в граммах:', int(df_day['Кол-во У'].sum()))
```

### Задача 3

Вася составил раскладку по продуктам на весь поход (она на листе "Раскладка") с указанием номера дня, названия продукта и его количества в граммах. Для каждого дня посчитайте 4 числа: суммарную калорийность и граммы белков, жиров и углеводов. Числа округлите до целых вниз и введите через пробел. Информация о каждом дне должна выводиться в отдельной строке.

```python
# Подключаем файл и заменяем NaN на 0, как в условии задачи
# Изменяем имя 1 столбца в справочнике для удобства
df_dir = pd.read_excel('trekking3.xlsx', sheet_name='Справочник').fillna(0)
df_dir.rename(columns={'Unnamed: 0':'Продукт'}, inplace=True)
df_day = pd.read_excel('trekking3.xlsx', sheet_name='Раскладка')

print(df_dir.head(10))
print(df_day)
```

```python
# Добавляем недостающую информацию из справочника в раскладку
df_day = df_day.merge(df_dir, on='Продукт')
df_day
```

```python
# Проводим необходимые подсчеты
df_day['Кол-во ККал'] = df_day['ККал на 100'] / 100 * df_day['Вес в граммах']
df_day['Кол-во Б'] = df_day['Б на 100'] / 100 * df_day['Вес в граммах']
df_day['Кол-во Ж'] = df_day['Ж на 100'] / 100 * df_day['Вес в граммах']
df_day['Кол-во У'] = df_day['У на 100'] / 100 * df_day['Вес в граммах']

# Удаляем лишние столбцы для удобства
df_day.drop(columns = ['ККал на 100', 'Б на 100', 'Ж на 100', 'У на 100', 'Вес в граммах'], inplace=True)
df_day
```

```python
# Группируем значения по дню и суммируем
df_day_sums = df_day.groupby(['День']).sum()
df_day_sums 
```

### Задача 4


Главный бухгалтер компании "Рога и копыта" случайно удалил ведомость с начисленной зарплатой. К счастью, у него сохранились расчётные листки всех сотрудников. Помогите по этим расчётным листкам восстановить зарплатную ведомость.

Ведомость должна содержать 1000 строк, в каждой строке должно быть указано ФИО сотрудника и, через пробел, его зарплата. Сотрудники должны быть упорядочены по алфавиту.

```python
# Проходимся циклом по xlsx, получаем значения фио и зп
# Записываем все в dataframe
df_zp = pd.DataFrame({'ФИО':None, 'Начислено':None}, index=[1])
for i in range(1, 1001):
    file = f'./2.4/{i}.xlsx'
    df1 = pd.read_excel(file)
    df_zp.loc[i] = [df1.iloc[0][1], df1.iloc[0][3]]

# Сортируем по алфавиту
df_zp = df_zp.sort_values(by='Начислено', ascending=False)
df_zp 
```
